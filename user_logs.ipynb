```python
def convert_date(x):
    x = str(x)
    date = datetime(year=int(x[0:4]), month=int(x[4:6]), day=int(x[6:8]))
    #date = datetime.strptime(date, "%Y-%m-%d")
    return date
user_loc["date"] = user_loc["date"].apply(lambda x: convert_date(x))
```

```python
user_total_logs = user_loc.groupby("msno").count()["total_secs"].to_frame()
user_total_logs.columns = ["nb_entries"]
```

```python
user_first_listen = user_loc.groupby("msno")["date"].min().to_frame()
user_first_listen.columns = ["first_listen"]
user_last_listen = user_loc.groupby("msno")["date"].max().to_frame()
user_last_listen.columns = ["last_listen"]
```

```python
user_first_last_listen = pd.concat([user_first_listen, user_last_listen], axis = 1)
```

```python
user_first_last_listen
```

```python
user_first_last_listen["length"] = (user_first_last_listen["first_listen"] - user_first_last_listen["last_listen"]).astype('timedelta64[D]')
```

```python
user_first_last_listen = pd.concat([user_first_last_listen, user_total_logs], axis = 1)
```
