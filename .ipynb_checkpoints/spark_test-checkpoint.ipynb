```python
# function that fill convert the date from 20160607 format into datetime
def convert_date(x):
    x = str(x)
    date = datetime(year=int(x[0:4]), month=int(x[4:6]), day=int(x[6:8]))
    return date
```

```python
from pyspark.sql.types import DateType
from pyspark.sql.functions import udf
```

```python
udf_myFunction = udf(convert_date, DateType())
```

```python
logs = logs.withColumn("date_timestamp", udf_myFunction("date"))
```
